```python
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
df
```

```python
dummies = pd.get_dummies(df['Car Model']).astype(int)
dummies
```

```python
mergeddf = pd.concat([df, dummies], axis='columns')
mergeddf
```

```python
# We have to drop Car Model, because we have dummy variables named after unique car models.
# We will drop one of the dummy variable to avoide dummy variable trap.
finaldf = mergeddf.drop(['Car Model','Audi A5'], axis='columns')
finaldf
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

X = finaldf.drop('Sell Price($)', axis='columns')
Y = df['Sell Price($)']
```

```python
reg.fit(X, Y)
```

```python
X_hat = pd.DataFrame([[90000, 8, 0, 0]], columns=['Mileage', 'Age(yrs)', 'BMW X5', 'Mercedez Benz C class'])
reg.predict(X_hat)
```
