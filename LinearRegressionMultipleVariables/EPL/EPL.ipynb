```python
import pandas as pd
from sklearn import linear_model
```

```python
df0 = pd.read_csv('EPLCSV.csv', sep='\t')
df0.to_csv('EPLdata.csv', sep=',', index=False)
df = pd.read_csv('EPLdata.csv')
df
```

```python
df.HomeTeam = df.HomeTeam.fillna('NA')
df.AwayTeam	 = df.AwayTeam	.fillna('NA')
df['FTH Goals'] = df['FTH Goals'].fillna(0)
df['FTA Goals'] = df['FTA Goals'].fillna(0)
df['FT Result'] = df['FT Result'].fillna('D')
df['Home_Team_Encoded'] = pd.factorize(df['HomeTeam'])[0]
df['Away_Team_Encoded'] = pd.factorize(df['AwayTeam'])[0]
df['FT_Result_Encoded'] = pd.factorize(df['FT Result'])[0]
df
```

```python
reg = linear_model.LinearRegression()
X = df[['Home_Team_Encoded','Away_Team_Encoded']]
Y = df[['FTH Goals','FTA Goals']]
reg.fit(X,Y)
```

```python
Home_Teams = df[['HomeTeam', 'Home_Team_Encoded']].drop_duplicates().reset_index(drop=True)
print("Home teams")
Home_Teams
```

```python
Away_Teams = df[['AwayTeam', 'Away_Team_Encoded']].drop_duplicates().reset_index(drop=True)
print("Away teams")
Away_Teams
```

```python
X_hat = pd.DataFrame([[15,10]], columns=['Home_Team_Encoded','Away_Team_Encoded'])
Y_hat = reg.predict(X_hat)
print(Y_hat)
```
