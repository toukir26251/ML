```python
# panda for dataframe
import pandas as pd
# for word to number conversion we need wodr2number
# pip install word2number
from word2number import w2n
from sklearn import linear_model
```

```python
# importing data in dataframe
df = pd.read_csv('hiring.csv')
df
```

```python
# filling zero to the emply row of experience column
df.experiece = df.experience.fillna('zero', inplace=True)
df
```

```python
# filling with mean value in NAN rows of test_score column
mean_test_score = df['test_score(out of 10)'].mean()
df['test_score(out of 10)'].fillna(mean_test_score, inplace=True)
df
```

```python
# converting word to number in experience column
df.experience=df.experience.apply(w2n.word_to_num)
df
```

```python
# training model with features
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
Y = df['salary($)']
reg = linear_model.LinearRegression()
reg.fit(X,Y)
```

```python
# predicting for one input
X_hat = pd.DataFrame([[3, 7, 10]], columns=['experience', 'test_score(out of 10)', 'interview_score(out of 10)'])
Y_hat = reg.predict(X_hat)
print(Y_hat)
```

```python
# predicting output for multiple input
future_candidate=pd.DataFrame({
    'experience':[2,12,0],
    'test_score(out of 10)':[9,10,3],
    'interview_score(out of 10)':[6,10,4]
    
})
salary_predict=reg.predict(future_candidate)
# adding salary_predict column and showing in table
future_candidate['salary_predict']=salary_predict
future_candidate
```
